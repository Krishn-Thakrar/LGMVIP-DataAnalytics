**Task-2 : Image to Pencil Sketch with Python**

**Importing Libraries**

```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

**Read Image**

```python
read_img = cv2.imread("SSR.jpg")
```

Show Image

```python
show_img = plt.imshow(read_img)
plt.axis(False)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

**Convert image from BGR to RGB**

```python
rgb_img = cv2.cvtColor(read_img,cv2.COLOR_BGR2RGB)
plt.imshow(rgb_img)
plt.axis(False)
plt.show()
```

Grey image

```python
grey_img = cv2.cvtColor(rgb_img,cv2.COLOR_BGR2GRAY)
show_img = plt.imshow(grey_img)
plt.axis(False)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

**Inverted grey image**

```python
inverted_grey_img = 255-grey_img
show_img = plt.imshow(inverted_grey_img)
plt.axis(False)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Blur image

```python
blur_img = cv2.GaussianBlur(inverted_grey_img, (21,21), 0)
show_img = plt.imshow(blur_img)
plt.axis(False)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Inverted blur image

```python
inverted_blur_img = 255 - blur_img
show_img = plt.imshow(inverted_blur_img)
plt.axis(False)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

**Pencil sketch image**

```python
pencil_sketch_img = cv2.divide(grey_img,inverted_blur_img,scale=256)
show_img = plt.imshow(pencil_sketch_img)
plt.axis(False)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

Save pencil sketch image

```python
cv2.imwrite('SSR.jpg',pencil_sketch_img)
```

Original and Pencil sketch image

```python
rgb_sketch = cv2.cvtColor(pencil_sketch_img,cv2.COLOR_BGR2RGB)
plt.subplot(1,2,1)
plt.title("Original image")
plt.imshow(rgb_img)
plt.axis(False)
plt.subplot(1,2,2)
plt.title("Pencil Sketch image")
plt.imshow(rgb_sketch)
plt.axis(False)
plt.show()
cv2.waitKey(0)
cv2.destroyAllWindows()
```
